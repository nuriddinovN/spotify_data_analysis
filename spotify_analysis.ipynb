```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mus_data=pd.read_json('spotify2021-2024.json')
```

```python
print(mus_data.columns)
```

<h1>Data clearning</h1>

```python
drop_list=['platform','conn_country','ip_addr','spotify_track_uri','episode_name','episode_show_name', 'spotify_episode_uri', 'audiobook_title','audiobook_uri', 'audiobook_chapter_uri', 'audiobook_chapter_title','incognito_mode']
mus_data=mus_data.drop(drop_list,axis=1)
```

```python
renamed={"master_metadata_track_name":'music',"master_metadata_album_artist_name":'artist',"master_metadata_album_album_name":'album'}
mus_data=mus_data.rename(columns=renamed)
```

```python
mus_data['minutes']=mus_data['ms_played']/60000
mus_data.drop(["ms_played"],axis=1)
```

<h1>Data analysis</h1>

```python


# Group by both 'music' and 'artist', count occurrences
most_played_music = mus_data.groupby(["music", "artist"]).size().reset_index(name="count")

most_played_music = most_played_music.sort_values(by="count", ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 7))
sns.barplot(x="count", y="music", data=most_played_music)
plt.xlabel("Number of Plays")
plt.ylabel("Song Name")
plt.title("Top 10 Most Played Songs")
plt.tight_layout()
plt.savefig('top_songs.png', dpi=300)
plt.show()

print('most played songs')
most_played_music.head()
```

```python
top_artists=mus_data.groupby(['artist']).size().sort_values(ascending=False)
#plot
colors = sns.color_palette("magma", len(top_artists.head()))

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_artists.head(), labels=top_artists.head().index, autopct="%1.1f%%", startangle=140, colors=colors)
plt.title("Top 10 Most Played Artists")
plt.tight_layout()
plt.savefig('top_artists.png', dpi=300)
plt.show()

print('Top artists')
top_artists.head()
```

```python
mus_data["date"] = pd.to_datetime(mus_data["ts"]).dt.date
most_music_day = mus_data.groupby("date")["minutes"].sum().sort_values(ascending=False)
#plot 
plt.figure(figsize=(10, 6))
sns.lineplot(x=most_music_day.index, y=most_music_day.values, marker="o", color="b")
plt.xlabel("Date")
plt.ylabel("Total Play Time (minutes)")
plt.title("Top 10 Most Played Days")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('listening_time_analysis.png', dpi=300)
plt.show()


print("music day: \n",most_music_day.head(1))
print('Musicless day\n',most_music_day.tail(1) )
```
